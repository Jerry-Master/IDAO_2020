# Checkpoints Comparation with Train

## 1. Llibraries i funció comparació

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def score(satellite_predicted_values, satellite_true_values): 
    # the division, addition and subtraction are pointwise 
    smape =  np.mean(np.abs((satellite_predicted_values - satellite_true_values) 
        / (np.abs(satellite_predicted_values) + np.abs(satellite_true_values))))
    return 100 * (1 - smape)
```

## 2. Importació del train i les prediccions del train

```python
raw = pd.read_csv("train.csv")[['x_sim', 'y_sim', 'z_sim', 'Vx_sim', 'Vy_sim', 'Vz_sim']].values
reals = pd.read_csv("train.csv")[['x', 'y', 'z', 'Vx', 'Vy', 'Vz']].values
improveds = pd.read_csv('submission.csv')[['x', 'y', 'z', 'Vx', 'Vy', 'Vz']].values
```

```python
if (len(improveds) != 649912):
    print("Warning: La mida del document hauria de ser 64991")
```

## 3. Puntuacions generals

```python
print("La puntuació del default  és:   " + str(score(raw, reals)))
print("La puntuació del improved és:   " + str(score(improveds, reals)))
```

## 4. Puntuacions particulars

```python
raw = pd.read_csv("train.csv")[['sat_id','x_sim', 'y_sim', 'z_sim', 'Vx_sim', 'Vy_sim', 'Vz_sim']]
real = pd.read_csv("train.csv")[['sat_id','x', 'y', 'z', 'Vx', 'Vy', 'Vz']]
improved = pd.read_csv('submission.csv')[['sat_id','x', 'y', 'z', 'Vx', 'Vy', 'Vz']]
```

```python
tot = 0
for i in range(0,600):
    raw_i = raw[raw['sat_id']==i].values
    real_i = real[real['sat_id']==i].values
    improved_i = improved[improved['sat_id']==i].values
    
    print("El satèl·lit: " + str(i) + " té puntuació del default  és: " + str(score(raw_i[:,1:7], real_i[:,1:7])))
    print("El satèl·lit: " + str(i) + " té puntuació del improved és: " + str(score(improved_i[:,1:7], real_i[:,1:7])))
    print("----------")
```

```python
tot = 0
scores = [[],[]]
good_scores = [[],[]]
bad_scores = [[],[]]

for i in range(0,600):
    real_i = real[real['sat_id']==i].values
    improved_i = improved[improved['sat_id']==i].values
    
    s = score(improved_i[:,1:7], real_i[:,1:7])
        
    if (s < 60):
        bad_scores[0].append(i)
        bad_scores[1].append(s)
        tot += 1
        print("El satèl·lit: " + str(i) + " té puntuació del improved és: " + str(score(improved_i[:,1:7], real_i[:,1:7])))
        print("----------")
        
    else:
        good_scores[0].append(i)
        good_scores[1].append(s)
    
    scores[0].append(i)
    scores[1].append(s)
    
```

## 5. Gràfics

```python
plt.figure(1)
plt.scatter(good_scores[0], good_scores[1], color = 'b')
plt.scatter(bad_scores[0], bad_scores[1], color = 'r')
plt.show()
```

```python
N = 600
theta = scores[1]
colors = (np.array(theta)/100)**20
plt.figure(2)
plt.plot([0, 600],[score(improveds, reals), score(improveds, reals)], 'r')
plt.scatter(scores[0], scores[1], c = colors)
plt.show()
```
